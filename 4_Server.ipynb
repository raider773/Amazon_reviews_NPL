```python
#host the "APP" with flask so the "Client" (the other notebook) gives the review to the model and model predicts
import pandas as pd
import numpy as np
import json
import pickle
from flask import  Flask, request, jsonify, render_template
```

```python
app = Flask('Sentiment_Analysis')
```

```python

@app.route('/api',methods=['POST'])
def predict_post():
   
    data = request.get_json(force=True)
    
    with open('model_binary.pkl', 'rb') as f_model:
        modelo = pickle.load(f_model)        
        m = modelo
        
           
    predictions = m.predict(data['review'])    
    
    
    return str(int(predictions[0]))
```

```python
app.run(host='0.0.0.0')
```
